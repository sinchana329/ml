## Part 2: Data Cleaning (Diabetes Dataset)

```python
diabetes = pd.read_csv('diabetes.csv')
diabetes.head()
```

```python
# Handling Missing Values
diabetes.fillna(diabetes.mean(), inplace=True)
diabetes.isnull().sum()
```

### Handling Outliers

```python
import numpy as np

Q1 = diabetes.quantile(0.25)
Q3 = diabetes.quantile(0.75)
IQR = Q3 - Q1

diabetes_clean = diabetes[~((diabetes < (Q1 - 1.5 * IQR)) | (diabetes > (Q3 + 1.5 * IQR))).any(axis=1)]
diabetes_clean.head()
```

## Part 3: Adult Income Dataset - Handling Categorical Data

## Part 4: Data Transformation

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Min-Max Normalization
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(diabetes_clean)
scaled_data[:5]
```

```python
# Standard Scaler
std_scaler = StandardScaler()
standardized_data = std_scaler.fit_transform(diabetes_clean)
standardized_data[:5]
```
