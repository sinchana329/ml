```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/ML Lab 3/hiring.csv")

print("Original data:\n", df)

# --- Word to number converter ---
word_to_num = {
    "zero": 0, "one": 1, "two": 2, "three": 3, "four": 4,
    "five": 5, "six": 6, "seven": 7, "eight": 8, "nine": 9,
    "ten": 10, "eleven": 11, "twelve": 12
}

# Convert experience column
df['experience'] = df['experience'].replace(word_to_num)

# Convert to numeric (force errors → NaN)
df['experience'] = pd.to_numeric(df['experience'], errors='coerce')

# Fill missing values
df['experience'] = df['experience'].fillna(0)

median_score = df['test_score(out of 10)'].median()
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(median_score)

print("\nCleaned data:\n", df)

# Features and target
X = df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']]
y = df['salary($)']

# Train model
model = LinearRegression()
model.fit(X, y)

# Predictions
pred1 = model.predict([[2, 9, 6]])
pred2 = model.predict([[12, 10, 10]])

print("\nPredicted salary (2 yr, 9 score, 6 interview):", pred1[0])
print("Predicted salary (12 yr, 10 score, 10 interview):", pred2[0])
```

```python
import pandas as pd
import numpy as np

# Load dataset
df = pd.read_csv("/content/drive/MyDrive/ML Lab 3/1000_Companies.csv")

print("Original data:\n", df.head())

# One-hot encode State
df = pd.get_dummies(df, columns=['State'], drop_first=True)

# Convert everything to float
df = df.astype(float)

# Split features and target
X = df.drop('Profit', axis=1).values
y = df['Profit'].values

# Add bias column
ones = np.ones((X.shape[0], 1))
X = np.hstack((ones, X))

# Normal Equation
theta = np.linalg.inv(X.T @ X) @ X.T @ y

print("\nModel parameters (theta):\n", theta)

# Prediction
new_data = np.array([1, 91694.48, 515841.3, 11931.24, 0, 1], dtype=float)

prediction = new_data @ theta

print("\nPredicted Profit:", prediction)
```
